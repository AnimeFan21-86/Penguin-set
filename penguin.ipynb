```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from palmerpenguins import load_penguins


df = load_penguins()
print(df.head())
mass_avg = df.groupby("sex")["body_mass_g"].mean() #filters out sex and then corresponding mass
print(mass_avg)
mass_avg.plot(kind='bar')
```

```python
 # bar chart of the mean bill length amongst all species 
df = load_penguins()
ade = df.groupby('species')['bill_length_mm'].mean()
print(ade)
ade.plot(kind='bar')
plt.ylabel('Mean bill length (mm)')
plt.title('Average bill length amongst penguin species')
```

```python
# scatter graphs of flipper length against body mass for female and male adelie species
ade_fe_fl = df[(df['species'] == 'Adelie') & (df['sex'] == 'female')]['flipper_length_mm']
ade_fe_bm = df[(df['species'] == 'Adelie') & (df['sex'] == 'female')]['body_mass_g']


ade_ma_fl = df[(df['species'] == 'Adelie') & (df['sex'] == 'male')]['flipper_length_mm']
ade_ma_bm = df[(df['species'] == 'Adelie') & (df['sex'] == 'male')]['body_mass_g']

m, b =np.polyfit(ade_fe_fl,ade_fe_bm,1) # first degree polynomial using the flipper length and body mass data
m1, b1 =np.polyfit(ade_ma_fl,ade_ma_bm,1)

plt.subplot(1,2,1)
plt.scatter(ade_fe_fl, ade_fe_bm, color = 'red')
plt.plot(ade_fe_fl, m*ade_fe_fl+b, 'black')
plt.xlabel('Flipper leßngth (mm)')
plt.ylabel('body mass (g)')
plt.title('Female ')

plt.subplot(1,2,2)
plt.scatter(ade_ma_fl, ade_ma_bm, color = 'blue')
plt.plot(ade_ma_fl, m1*ade_ma_fl+b1, 'black')
plt.xlabel('Flipper leßngth (mm)')
plt.ylabel('body mass (g)')
plt.title('Male')
plt.show()
```

```python
m_fem, b_fem =np.polyfit(ade_fe_fl,ade_fe_bm,1) # first degree polynomial using the flipper length and body mass data
m_mal, b_mal =np.polyfit(ade_ma_fl,ade_ma_bm,1)

fig, axs = plt.subplots(nrows=1, ncols=2, sharex=True, sharey=True)
fig.set_size_inches(6,2.5)
fig.subplots_adjust(wspace=0)
fig.set_dpi(225)

#Plotting happens here
axs[0].scatter(ade_fe_fl, ade_fe_bm, color = 'red')
#axs[0].plot(ade_fe_fl, m_fem*ade_fe_fl+b_fem, 'black')

axs[0].scatter(ade_ma_fl, ade_ma_bm, color = 'blue')
#axs[0].plot(ade_ma_fl, m_mal*ade_ma_fl+b_mal, 'black')

axs[1].scatter(ade_fe_fl, ade_fe_bm, color = 'green')

axs[1].scatter(ade_ma_fl, ade_ma_bm, color = 'green')
#Plotting stops here

axs[0].set_ylabel('Mass (g)')
fig.text(0.425, -0.005, 'Flipper Length (cm)')
```

```python
# exploring the differences in fliffper length and body mass as a means of differentating species

# Species
ade_fl = df[(df['species'] == 'Adelie')]['flipper_length_mm']
chi_fl = df[(df['species'] == 'Chinstrap')]['flipper_length_mm']
gen_fl = df[(df['species'] == 'Gentoo')]['flipper_length_mm']

# body mass
ade_bm = df[(df['species'] == 'Adelie')]['body_mass_g']
chi_bm = df[(df['species'] == 'Chinstrap')]['body_mass_g']
gen_bm = df[(df['species'] == 'Gentoo')]['body_mass_g']


fig, axs = plt.subplots(nrows=1, ncols=2, sharex=True, sharey=True)
fig.set_size_inches(6,2.5)
fig.subplots_adjust(wspace=0)
fig.set_dpi(225)

#Plotting happens here
axs[0].scatter(ade_fl, ade_bm, color = 'red',  label = 'Adelie')
#axs[0].plot(ade_fe_fl, m_fem*ade_fe_fl+b_fem, 'black')

axs[0].scatter(chi_fl, chi_bm, color = 'blue', label = 'Chinstrip')
#axs[0].plot(ade_ma_fl, m_mal*ade_ma_fl+b_mal, 'black')

axs[0].scatter(gen_fl, gen_bm, color = 'green', label = 'Gentoo')
#axs[0].plot(ade_ma_fl, m_mal*ade_ma_fl+b_mal, 'black')

axs[1].scatter(ade_fl, ade_bm, color = 'k')

axs[1].scatter(chi_fl, chi_bm, color = 'k')

axs[1].scatter(gen_fl, gen_bm, color = 'k')
#Plotting stops here

axs[0].set_ylabel('Mass (g)')
fig.text(0.425, -0.005, 'Flipper Length (cm)')
fig.legend()

```
