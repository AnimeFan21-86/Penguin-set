```python
# imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from palmerpenguins import load_penguins
df = load_penguins()
print(df.head())
```

Convecntions to follow:

Red = Adelie
Green = Gentoo
Blue = chinstrap

```python
# exploring the differences in fliffper length and body mass as a means of differentating species

# Species
ade_fl = df[(df['species'] == 'Adelie')]['flipper_length_mm']
chi_fl = df[(df['species'] == 'Chinstrap')]['flipper_length_mm']
gen_fl = df[(df['species'] == 'Gentoo')]['flipper_length_mm']

# body mass
ade_bm = df[(df['species'] == 'Adelie')]['body_mass_g']
chi_bm = df[(df['species'] == 'Chinstrap')]['body_mass_g']
gen_bm = df[(df['species'] == 'Gentoo')]['body_mass_g']


fig, axs = plt.subplots(nrows=1, ncols=2, sharex=True, sharey=True)
fig.set_size_inches(6,2.5)
fig.subplots_adjust(wspace=0)
fig.set_dpi(225)

#Plotting happens here
axs[0].scatter(ade_fl, ade_bm, color = 'red',  label = 'Adelie')
#axs[0].plot(ade_fe_fl, m_fem*ade_fe_fl+b_fem, 'black')

axs[0].scatter(chi_fl, chi_bm, color = 'blue', label = 'Chinstrip')
#axs[0].plot(ade_ma_fl, m_mal*ade_ma_fl+b_mal, 'black')

axs[0].scatter(gen_fl, gen_bm, color = 'green', label = 'Gentoo')
#axs[0].plot(ade_ma_fl, m_mal*ade_ma_fl+b_mal, 'black')

axs[1].scatter(ade_fl, ade_bm, color = 'k')

axs[1].scatter(chi_fl, chi_bm, color = 'k')

axs[1].scatter(gen_fl, gen_bm, color = 'k')
#Plotting stops here

axs[0].set_ylabel('Mass (g)')
fig.text(0.425, -0.005, 'Flipper Length (cm)')
fig.legend()
plt.title('mass and flipper length')


```

```python
# exploring the bill length and bill depth as a means of identifying species

ade_fl = df[(df['species'] == 'Adelie')]['flipper_length_mm']
chi_fl = df[(df['species'] == 'Chinstrap')]['flipper_length_mm']
gen_fl = df[(df['species'] == 'Gentoo')]['flipper_length_mm']

# bill length
ade_bl = df[(df['species'] == 'Adelie')]['bill_length_mm']
chi_bl = df[(df['species'] == 'Chinstrap')]['bill_length_mm']
gen_bl = df[(df['species'] == 'Gentoo')]['bill_length_mm']

fig, axs = plt.subplots(nrows=1, ncols=2, sharex=True, sharey=True)
fig.set_size_inches(6,2.5)
fig.subplots_adjust(wspace=0)
fig.set_dpi(225)

axs[0].scatter(ade_fl,ade_bl,color = 'red', label = 'Adelie')
axs[0].scatter(chi_fl,chi_bl, color = 'blue', label = ' Chinstrip')
axs[0].scatter(gen_fl,gen_bl, color = 'green', label = 'Gentoo')

axs[1].scatter(ade_fl, ade_bl, color = 'k')
axs[1].scatter(chi_fl, chi_bl, color = 'k')
axs[1].scatter(gen_fl, gen_bl, color = 'k')

axs[0].set_ylabel('Bill length (mm))')
fig.text(0.425, -0.005, 'Flipper Length (cm)')
fig.legend()
plt.title('bill and flipper length')
```

```python
# ['Adelie' 'Gentoo' 'Chinstrap']
# ['Torgersen' 'Biscoe' 'Dream']

ade_tor = df[(df['species'] == 'Adelie') & (df['island'] == 'Torgersen')] # live in torgersen
ade_bi = df[(df['species'] == 'Adelie') & (df['island'] == 'Biscoe')] # live in biscoe
ade_dr = df[(df['species'] == 'Adelie') & (df['island'] == 'Dream')] # live in dream

chi_tor = df[(df['species'] == 'Chinstrap') & (df['island'] == 'Torgersen')] 
chi_bi = df[(df['species'] == 'Chinstrap') & (df['island'] == 'Biscoe')]
chi_dr = df[(df['species'] == 'Chinstrap') & (df['island'] == 'Dream')] 

gen_tor = df[(df['species'] == 'Gentoo') & (df['island'] == 'Torgersen')] 
gen_bi = df[(df['species'] == 'Gentoo') & (df['island'] == 'Biscoe')] 
gen_dr = df[(df['species'] == 'Gentoo') & (df['island'] == 'Dream')]

print(f"there are {len(ade_tor)} adelie in torgersen")
print(f"there are {len(ade_bi)} adelie in biscoe")
print(f"there are {len(ade_dr)} adelie in dream")

print(f"there are {len(chi_tor)} chinstrap in torgersen")
print(f"there are {len(chi_bi)} chinstrap in biscoe")
print(f"there are {len(chi_dr)} chinstrap in dream")

print(f"there are {len(gen_tor)} gentoo in torgersen")
print(f"there are {len(gen_bi)} gentoo in biscoe")
print(f"there are {len(gen_dr)} gentoo in dream")

# stacked bar charts
# the scores exist in order of their respecitve island
island = ['Torgersen', 'Biscoe' ,'Dream']
adelie_score = np.array([52, 44, 56])
chinstrap_score = np.array([0,0,68])
gentoo_score = np.array([0,124,0])

#plot
plt.bar(island, adelie_score, color = 'red')
plt.bar(island, chinstrap_score, bottom= adelie_score, color = 'blue')
plt.bar(island,gentoo_score,bottom= adelie_score + chinstrap_score, color = 'green')
plt.xlabel('Island')
plt.ylabel('Frequnecy')
plt.legend(['Adelie', 'Chinstrap','Gentoo'])
plt.show()







```

```python
# lets consider the distribution of mass amongst each species

# body mass
ade_bm = df[(df['species'] == 'Adelie')]['body_mass_g']
chi_bm = df[(df['species'] == 'Chinstrap')]['body_mass_g']
gen_bm = df[(df['species'] == 'Gentoo')]['body_mass_g']


# plot histo
plt.hist(ade_bm, color= 'red') 
plt.hist(chi_bm, color= 'blue')
plt.hist(gen_bm, color='green')
plt.legend(['Adelie', 'Chinstrap','Gentoo'])
plt.xlabel('mass (g)')
plt.ylabel('frequnecy')
plt.title('Distribution of mass amongst species')

```

```python
# boxplot of masses amonst specie for better idea
# body mass
ade_bm = df[(df['species'] == 'Adelie')]['body_mass_g']
chi_bm = df[(df['species'] == 'Chinstrap')]['body_mass_g']
gen_bm = df[(df['species'] == 'Gentoo')]['body_mass_g']

sns.boxplot(data= [ade_bm, chi_bm , gen_bm])

plt.xticks([0, 1, 2], ['Adelie', 'Chinstrap', 'Gentoo'])
```

```python
# 3 dimesnioal scatter plot of mass, flipper and bill length

# mass 
ade_bm = df[(df['species'] == 'Adelie')]['body_mass_g']
chi_bm = df[(df['species'] == 'Chinstrap')]['body_mass_g']
gen_bm = df[(df['species'] == 'Gentoo')]['body_mass_g']

# flipper
ade_fl = df[(df['species'] == 'Adelie')]['flipper_length_mm']
chi_fl = df[(df['species'] == 'Chinstrap')]['flipper_length_mm']
gen_fl = df[(df['species'] == 'Gentoo')]['flipper_length_mm']

# bill
ade_bl = df[(df['species'] == 'Adelie')]['bill_length_mm']
chi_bl = df[(df['species'] == 'Chinstrap')]['bill_length_mm']
gen_bl = df[(df['species'] == 'Gentoo')]['bill_length_mm']

fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')

ax.scatter3D(ade_bm, ade_fl, ade_bl, color='red', marker='o')
ax.scatter3D(chi_bm, chi_fl, chi_bl, color='blue', marker='o')
ax.scatter3D(gen_bm, gen_fl, gen_bl, color='green', marker='o')



# Labels
ax.set_xlabel('body mass (g)')
ax.set_ylabel('flipper length (mm) ')
ax.set_zlabel('bill length (mm)')
ax.set_title('3D Scatter Plot of mass, flipper and bill length')


plt.show()
```
