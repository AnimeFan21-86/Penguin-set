```python
# imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm
import statistics as stat

from palmerpenguins import load_penguins
df = load_penguins()
print(df.head())
```

Convecntions to follow:

Red = Adelie
Green = Gentoo
Blue = chinstrap

```python
# exploring the differences in fliffper length and body mass as a means of differentating species

# Species
ade_fl = df[(df['species'] == 'Adelie')]['flipper_length_mm']
chi_fl = df[(df['species'] == 'Chinstrap')]['flipper_length_mm']
gen_fl = df[(df['species'] == 'Gentoo')]['flipper_length_mm']

# body mass
ade_bm = df[(df['species'] == 'Adelie')]['body_mass_g']
chi_bm = df[(df['species'] == 'Chinstrap')]['body_mass_g']
gen_bm = df[(df['species'] == 'Gentoo')]['body_mass_g']


fig, axs = plt.subplots(nrows=1, ncols=2, sharex=True, sharey=True)
fig.set_size_inches(6,2.5)
fig.subplots_adjust(wspace=0)
fig.set_dpi(225)

#Plotting happens here
axs[0].scatter(ade_fl, ade_bm, color = 'red',  label = 'Adelie')
#axs[0].plot(ade_fe_fl, m_fem*ade_fe_fl+b_fem, 'black')

axs[0].scatter(chi_fl, chi_bm, color = 'blue', label = 'Chinstrip')
#axs[0].plot(ade_ma_fl, m_mal*ade_ma_fl+b_mal, 'black')

axs[0].scatter(gen_fl, gen_bm, color = 'green', label = 'Gentoo')
#axs[0].plot(ade_ma_fl, m_mal*ade_ma_fl+b_mal, 'black')

axs[1].scatter(ade_fl, ade_bm, color = 'k')

axs[1].scatter(chi_fl, chi_bm, color = 'k')

axs[1].scatter(gen_fl, gen_bm, color = 'k')
#Plotting stops here

axs[0].set_ylabel('Mass (g)')
fig.text(0.425, -0.005, 'Flipper Length (cm)')
fig.legend()
plt.title('mass and flipper length')


```

```python
# exploring the bill length and bill depth as a means of identifying species

ade_fl = df[(df['species'] == 'Adelie')]['flipper_length_mm']
chi_fl = df[(df['species'] == 'Chinstrap')]['flipper_length_mm']
gen_fl = df[(df['species'] == 'Gentoo')]['flipper_length_mm']

# bill length
ade_bl = df[(df['species'] == 'Adelie')]['bill_length_mm']
chi_bl = df[(df['species'] == 'Chinstrap')]['bill_length_mm']
gen_bl = df[(df['species'] == 'Gentoo')]['bill_length_mm']

fig, axs = plt.subplots(nrows=1, ncols=2, sharex=True, sharey=True)
fig.set_size_inches(6,2.5)
fig.subplots_adjust(wspace=0)
fig.set_dpi(225)

axs[0].scatter(ade_fl,ade_bl,color = 'red', label = 'Adelie')
axs[0].scatter(chi_fl,chi_bl, color = 'blue', label = ' Chinstrip')
axs[0].scatter(gen_fl,gen_bl, color = 'green', label = 'Gentoo')

axs[1].scatter(ade_fl, ade_bl, color = 'k')
axs[1].scatter(chi_fl, chi_bl, color = 'k')
axs[1].scatter(gen_fl, gen_bl, color = 'k')

axs[0].set_ylabel('Bill length (mm))')
fig.text(0.425, -0.005, 'Flipper Length (cm)')
fig.legend()
plt.title('bill and flipper length')
```

```python
# lets consider the distribution of mass amongst each species

# body mass
ade_bm = df[(df['species'] == 'Adelie')]['body_mass_g']
chi_bm = df[(df['species'] == 'Chinstrap')]['body_mass_g']
gen_bm = df[(df['species'] == 'Gentoo')]['body_mass_g']


# plot histo, bins based on sqrt of N points
plt.hist(ade_bm,bins = 20,color= 'red') 
plt.hist(chi_bm,bins = 20,color= 'blue')
plt.hist(gen_bm,bins = 20,color='green')
plt.legend(['Adelie', 'Chinstrap','Gentoo'])
plt.xlabel('mass (g)')
plt.ylabel('frequnecy')
plt.title('Distribution of mass amongst species')

print(df[(df['species'] == 'Adelie')]['body_mass_g'].max()) # min:2850 max: 4775
print(df[(df['species'] == 'Chinstrap')]['body_mass_g'].max()) # 2700 max: 4800
print(df[(df['species'] == 'Gentoo')]['body_mass_g'].max()) #3950 max: 6300

# mean 
#print(np.mean(ade_bm)) # 3700
#print(np.mean(chi_bm)) # 3733
#print(np.mean(gen_bm)) # 5076

# standard deviation
#print(np.std(ade_bm)) #457
#print(np.std(chi_bm)) #381
#print(np.std(gen_bm)) # 502






```

```python
# calc x axis range
x_axis_ade = np.arange(2850,4775,300)
x_axis_chi = np.arange(2700,4800,300)
x_axis_gen = np.arange(3950,6300,300)

# calc mean and std for each species alongside their pdf
mu_ade, std_ade = norm.fit(ade_bm.dropna())
p_ade= norm.pdf(x_axis_ade,mu_ade,std_ade)

mu_chi, std_chi = norm.fit(chi_bm.dropna())
p_chi = norm.pdf(x_axis_chi,mu_chi,std_chi)

mu_gen, std_gen = norm.fit(gen_bm.dropna())
p_gen = norm.pdf(x_axis_gen,mu_gen,std_gen)

# subplot of figs

fig, axs = plt.subplots(nrows=1, ncols=4, sharex=True, sharey=True)
fig.set_size_inches(6,2.5)
fig.subplots_adjust(wspace=0)
fig.set_dpi(225)

axs[0].hist(ade_bm,bins = 20,density=True,color= 'red', label = 'adelie')
axs[0].plot(x_axis_ade,p_ade,'k',linewidth =2)

axs[1].hist(chi_bm,bins = 20,density=True,color= 'blue', label = 'chinstrap')
axs[1].plot(x_axis_chi,p_chi,'k',linewidth =2)

axs[2].hist(gen_bm,bins = 20,density=True,color= 'green', label = 'gentoo')
axs[2].plot(x_axis_gen,p_gen,'k',linewidth =2)

# 4th colum misclenaious
axs[3].hist(ade_bm,bins = 20,density=True,color= 'red')
axs[3].plot(x_axis_ade,p_ade,'k',linewidth =2)

axs[3].hist(chi_bm,bins = 20,density=True,color= 'blue')
axs[3].plot(x_axis_chi,p_chi,'k',linewidth =2)

axs[3].hist(gen_bm,bins = 20,density=True,color= 'green')
axs[3].plot(x_axis_gen,p_gen,'k',linewidth =2)

axs[0].set_ylabel('frequnecy ')
fig.text(0.425, -0.005, 'mass')
fig.legend()
```
